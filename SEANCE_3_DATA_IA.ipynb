```python
#Ecrire une fonction qui reçoit en entrée une valeur et qui retourne un intervalle

def var_cat(valeur):
    if valeur < 18:
        valeur = " < 18ans "
    elif (18<=valeur<35):
        valeur="18-35ans"
    elif (35<=valeur<60):
        valeur="35-60ans"
    else :
        valeur= ">=60ans"
    return valeur
```

```python
var_cat(20)
```

```python
#recoder la variable age

dataset['Age'] = dataset['Age'].map(var_cat) 
#  dataset.apply(var_cat) , apply ne marche qu'avec les variables quantitatives
#il est donc plus judicieux d'utiliser map
```
