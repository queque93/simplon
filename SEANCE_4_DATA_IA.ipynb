```python
#Obtenir la colonne presentant la plus grande donnÃ©e manquante

df.isna().sum().nlargest(3)
```

```python
#remplacer la valeur manquante de pression par la pression_moyenne

df['pression'].fillna(df['pression'].mean(),inplace=True) # inplace pour garder en memoire 
```

```python
df['pression'].isna().any()
```

```python
#imputation sur plusieurs variables

df.fillna({'cholester':df['cholester'].mean(), 'pression':df['pression'].mean(),})
```

```python
#Tri a plat de la variable angine

df['angine'].value_counts()
```

```python
#imputer par la moyenne et le mode 

df.fillna({'cholester':df['cholester'].mean(), 'pression':df['pression'].mean(), 'angine':'non'},inplace=True)
```

```python
df.isna().sum()
```

```python
#supprimer la colonne angine

df.drop(["angine"], axis=1)         # del df['angine'] 
```

```python
df
```

```python
#supprimer les lignes qui contiennent les valeurs manquantes suivant la variable cholester

df.dropna(axis=0,subset=['cholester'],inplace=True)
```

```python
df
```

```python
#Donner le nombre de femmes et d'hommes

df['sexe'].value_counts()
```

```python
nbre_masc=(df['sexe']=='masculin').sum()
nbre_fem=(df['sexe']=='feminin').sum()
```

```python
#Donner le tableau de contingence des personnes presentant une maladie cardiaque ou pas selon le sexe

tb=pd.crosstab(df['sexe'],df['coeur'])
tb
```

```python
#le type du tableau

type(tb)
```

```python
#calculer un taux d'erreur

(tb.iloc[0,1]+tb.iloc[1,0])/(tb.sum().sum())*100
```
